```python
import numpy as np
import matplotlib.pyplot as plt
import radial  # Ensure radial.py is available
import radiallog  # Ensure radiallog.py is available

# Configurations for the orbitals
Z = 1  # Atomic number
orbitals = [1, 2, 3, 6, 9]  # Principal quantum numbers

# Function to calculate the radial expectation value
def r_expectation_value(r, P):
    dr = r[1] - r[0]  # Linear spacing
    return np.sum(r * P**2) * dr

# Function to plot vertical lines for expectation values
def vlines(x_value, top_value, label, color):
    plt.vlines(x=x_value, ymin=0, ymax=top_value, colors=color, linestyles=':', label=label)

# Prepare plot
plt.figure(figsize=(10, 6))

# Loop through each orbital and plot
for n in orbitals:
    # radial.py data
    r, P, _ = radial.radial(0, n, Z)
    r_expectation_radial = r_expectation_value(r, P)

    # radiallog.py data
    rho, P_log, _ = radiallog.radiallog(0, n, Z)
    transformed_P_log = P_log * np.sqrt(rho)  # Apply Equation (12)
    r_expectation_radiallog = r_expectation_value(rho, transformed_P_log)

    # Plot radial.py wavefunction
    plt.plot(r, r * P**2, label=f'{n}s (radial.py)')

    # Plot radiallog.py wavefunction
    plt.plot(rho, rho * transformed_P_log**2, '--', label=f'{n}s (radiallog.py)')

    # Add vertical lines for radial expectation values
    vlines(r_expectation_radial, max(r * P**2), f'⟨r⟩ radial.py {n}s', color='blue')
    vlines(r_expectation_radiallog, max(rho * transformed_P_log**2), f'⟨r⟩ radiallog.py {n}s', color='green')

# Plot customization
plt.xlabel('r (Distance from nucleus)')
plt.ylabel(r'Radial Density Function $rP^2(r)$')
plt.title('Radial Density Functions and Radial Expectation Values')
plt.legend()
plt.grid()
plt.show()
```

```python
# Modified radiallog.py
# Program for solving the radial Schrödinger equation using a parametric potential
# V_alpha(r) = -Z/r + ((Z - zeta) * r) / (a^2 + r^2)

import math
import numpy as np
import matplotlib.pyplot as plt
import sys
import datetime

def radiallog(l, n, Z, a=0.2683, zeta=1, plot=True):
    starttime = datetime.datetime.now().timestamp()

    if Z > 0 and l < n:
        E = - Z**2 / 2  # starting energy
        nodes = n - l - 1  # targeted number of nodes
    else:
        sys.stderr.write('Invalid input\n')
        sys.exit(-1)

    rhomin = -10
    h = 1 / 48  # step size
    nodes_count = -1
    dE = math.inf
    E_upper = -2.0e-52
    E_lower = -math.inf
    num_iter = 0

    while nodes != nodes_count or abs(dE) > 1.0e-10:
        # Ensure turning point is within grid
        while True:
            r_inf = 40 / math.sqrt(2 * abs(E))
            rhoi = math.log(Z * r_inf)
            grid_points = round((-rhomin + rhoi) / h)
            rho = rhomin + np.linspace(0, grid_points - 1, grid_points) * h
            r = np.exp(rho) / Z

            # Define the effective potential U
            U = np.zeros(grid_points)
            U[1:] = (-Z / r[1:]) + ((Z - zeta) * r[1:]) / (a**2 + r[1:]**2) + l * (l + 1) / (2 * r[1:]**2)

            # Determine outer classical turning point
            i_c = grid_points - 1
            for i in range(grid_points - 1, -1, -1):
                if U[i] < E:
                    i_c = i
                    break

            if i_c == grid_points - 1:
                E = 0.9 * E
            else:
                break

        # Define the g function for Numerov method
        g = -2 * r**2 * (E + Z / r - (Z - zeta) * r / (a**2 + r**2)) + (l + 1 / 2)**2
        alpha = 1 - (h**2 / 12) * g
        beta = 2 + (5 * h**2 / 6) * g

        # Numerov integration
        psi = np.zeros(grid_points)
        psi[0] = 0
        psi[1] = 1e-10

        for i in range(1, i_c):
            psi[i + 1] = (beta[i] * psi[i] - alpha[i - 1] * psi[i - 1]) / alpha[i + 1]

        # Normalize the wavefunction
        norm = np.trapz(psi**2, r)
        psi /= np.sqrt(norm)

        # Count nodes
        nodes_count = np.sum(np.diff(np.sign(psi)) != 0)

        # Adjust energy based on node count
        if nodes_count > nodes:
            E_upper = E
        else:
            E_lower = E

        E = (E_upper + E_lower) / 2
        dE = E_upper - E_lower

        num_iter += 1

    if plot:
        plt.plot(r, psi, label=f'l={l}, n={n}')
        plt.xlabel('r (a.u.)')
        plt.ylabel('Radial Wavefunction')
        plt.title('Radial Wavefunction using Parametric Potential')
        plt.legend()
        plt.show()

    endtime = datetime.datetime.now().timestamp()
    print(f'Energy: {E:.6f} a.u., Iterations: {num_iter}, Time: {endtime - starttime:.2f}s')
    return E

radiallog (0,1,11)
```

```python
import math
import numpy as np
import matplotlib.pyplot as plt
import sys
import datetime

def radiallog(l,n,Z,a=0.2683, zeta=1,plot=True):

    starttime=datetime.datetime.now().timestamp()

    if Z > 0 and l < n:
        E = - Z**2/2                                      #starting energy taken as the lowest possible
        nodes = n - l - 1                                 #targeted number of nodes
    else:
        sys.stderr.write('Invalid input\n')
        sys.exit(-1)

    rhomin = -10
    h = 1/48                                              #step size
    nodes_count = -1                                      #initial values of nodes
    dE = math.inf                                         #initial values of difference of derivat at r_c
    E_upper = -2.0e-52                                    #initial values of upper bound of energy during iter
    E_lower = -math.inf                                   #initial values of lower bound of energy during iter
    num_iter = 0

    while nodes != nodes_count or abs(dE) > 1.0e-10:
        #Make sure that the energy is such that the turning point is in the interior of the grid
        while 1:
            r_inf = 40/math.sqrt(2*abs(E))                              #practical infinity
            rhoi = math.log(Z*r_inf);                                   # practical infinity in rho
            grid_points = round((-rhomin + rhoi)/h);                    # number of grid points
            rho = rhomin + np.linspace(0,grid_points-1,grid_points)*h;  # generate grid
            r = np.exp(rho)/Z;
            # Define the effective potential
            U = np.linspace(0, 0, grid_points)            #initialize U
            U[0] = 0                                      #gets a  0  value at r = 0
            U[1:] = (-Z / r[1:]) + ((Z - zeta) * r[1:]) / (a**2 + r[1:]**2) + l * (l + 1) / (2 * r[1:]**2)

            #Determine the outer classical turning point. Start from the practical infinity
            # and step inwards until U(i) < E
            i_c = grid_points-1
            for i in range(grid_points-1, -1, -1):
                if U[i] < E:
                    i_c = i
                    break
            #If turning point not in the interior increase the energy and try again
            if i_c == grid_points-1:
                E = 0.9*E
            else:
                break

        g = -2 * r**2 * (E + Z / r - (Z - zeta) * r / (a**2 + r**2)) + (l + 1 / 2)**2               #g function for Numerow
        alpha = 1-(h**2/12)*g                             #alpha and
        beta = 2+(5*h**2/6)*g                             #beta for Numerov's method

        #Perform the outward integration
        P_out = np.zeros(grid_points)                               #initialize P_out
        P_out[0] = r[0]**(l+1)*(1-Z*r[0]/(l+1))/math.sqrt(r[0])     #starting values from regular
        P_out[1] = r[1]**(l+1)*(1-Z*r[1]/(l+1))/math.sqrt(r[1])     #solution at r = 0
        for i in range(1,i_c+1):
            P_out[i+1] = (beta[i]*P_out[i] - alpha[i-1]*P_out[i-1])/alpha[i+1]

        #Perform the inward integration
        P_in = np.zeros(grid_points)                                                                    #initialize P_in
        P_in[grid_points-1] = np.exp(-np.sqrt(2*abs(E)*r[grid_points-1]))/math.sqrt(r[grid_points-1])   #starting values from
        P_in[grid_points-2] = np.exp(-np.sqrt(2*abs(E)*r[grid_points-2]))/math.sqrt(r[grid_points-2])   #regular solution at r_inf
        for i in range(grid_points-2,i_c-1,-1):
            P_in[i-1] = (beta[i]*P_in[i] - alpha[i+1]*P_in[i+1])/alpha[i-1]

        #Scale and merge solutions
        P_out = P_out/P_out[i_c]
        P_in = P_in/P_in[i_c]
        P = np.append(P_out[0:i_c], P_in[i_c:])

        #Count the number of nodes. At a node position, P(i+1)P(i) is negative
        nodes_count = 0
        for i in range(1,grid_points-1):
            if P[i+1]*P[i] < 0:
                nodes_count +=1

        #Check nodes and adjust E
        if nodes_count > nodes:                           #too many nodes, decrease E
            E_upper = min(E_upper,E)
            if E*1.1 < E_lower:
                E = (E_lower+E)/2
            else:
                E = E*1.1
        elif nodes_count < nodes:                         #too few nodes, increase E
            E_lower = max(E_lower,E)
            if E*0.9 > E_upper:
                E = (E+E_upper)/2
            else:
                E = E*0.9

        #Fine tune the energy
        if nodes_count == nodes:
            #Compute the derivatives and take difference
            DP_out = (P_out[i_c+1]-P_out[i_c-1])/(2*h)
            DP_in = (P_in[i_c+1]-P_in[i_c-1])/(2*h)
            dE = (DP_out-DP_in)*P[i_c]/(2*sum(r**2*P**2)*h)
            #Damp if step is too large
            while E+dE > 0:
                dE = dE/2
            E = E + dE
        print("Iteration %d: E = %.16e" %(num_iter,E,))
        num_iter += 1

    #Normalize and check slope
    norm = math.sqrt(sum(P**2*r**2)*h)
    P = P/norm
    if P[1] < 0:
        P = -P

    #Compute radial expectation value
    r1 = sum(r**3*P**2)*h

    endtime=datetime.datetime.now().timestamp()
    print()
    print("time used: %f s" %(endtime-starttime))

    #Write energy and plot solution
    print()
    print("Number of iterations: %d" %(num_iter,))
    print("Number of grid points: %d" %(grid_points,))
    print("Energy eigenvalue: %.16e a.u." %(E,))
    print("Radial expectation value: %.16e a.u." %(r1,))

    orbital=["s","p","d","f","g","h","i"]

    if plot:
        plt.figure()
        plt.subplot(211)
        plt.plot(r,np.sqrt(r)*P)
        plt.xlabel("r (a.u.)")
        plt.title("radial function for %d%s of Z = %d" %(n,orbital[l],Z,))
        plt.ylabel("P")
        rr = np.sqrt(r)
        plt.subplot(212)
        plt.plot(rr,np.sqrt(r)*P)
        plt.xlabel('$\sqrt{\mathrm{r}}$ (a.u.)')
        plt.ylabel("P")
        plt.tight_layout(pad=1.0)
        plt.show()

    return (r,P,E,)

if __name__ == '__main__':
    #input
    l = int(input("Give the orbital quantum number l: "))
    n = int(input("Give the principal quantum number n: "))
    Z = int(input("Give the nuclear charge Z: "))
    print()
    radiallog(l,n,Z)


```
